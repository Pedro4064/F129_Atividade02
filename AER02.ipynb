# Questão 1

```python
def NPS(Pef):
    Pref = 2 * (10**-5)
    return 20 * math.log((Pef/Pref),10)

def Pef(NPS):
    Pref = 2 * (10**-5)
    return 10**((NPS/20) + math.log(Pref, 10))
```

```python
print('Letra a ->', NPS(0.36))
print('Letra b ->', NPS(22))
print('Letra c ->', Pef(95))
```

# Questão 2
Considerando que para liearizarmos $y = cx^d$ para o formato $m=k+dn$ temos:

$$\begin{cases}m = \log(y) \\ k = log(c) \\ n = log(x)\end{cases}$$

Tendo isso em mente, para linearizarmos: 
$$y = (9x)^{\frac{1}{2}} = 9^{\frac{1}{2}} x^{\frac{1}{2}}$$

teremos: 
$$\begin{cases}m = \log(y) \\ k = \frac{1}{2}log(9) \\ n = log(x) \\ d = \frac{1}{2}\end{cases}$$

```python
# Para calcularmos o valor de k
print('Valor de k->',np.round(math.log(9,10)/2, 1))
```

# Questão 4

Gráfico que corresponde à função: 
$$y=x^3$$
No papel $\log\times \log$.

Primeiramente precisamos linearizar, como foi feito na questão 2, com isso temos: 
$$
\begin{cases}
m = log(y)\\
n = log(x)\\
k=log(1) = 0\\
d = 3\\
\end{cases}
$$

resultando em uma função linearizada: 
$$m = 0+3n$$

Podemos plotar em escala $\log\times \log$ para podermos ver que realmente acabamos com uma reta descrita acima.

```python
def m(x):
    return x**3

x = np.linspace(1,100,10)
y = m(x)

plt.plot(x,y)
plt.grid(which='both')
plt.yscale('log')
plt.xscale('log')
plt.show()
```

# Questão 7
Sabendo que a fórmula para pressão em função do volume(para gases ideais) é: 
$$P = \frac{nRT}{V}$$

Onde temos:
$$n=1mol\\ R=8.14 \\ T=300K$$

Temos o seguinte gráfico (alternativa 2):


```python
def P(v):
    n = 1 
    R = 8.14 
    T = 300
    return (n*R*T)/v
    # return v**-1

x = np.linspace(1,1000,1000)
y = P(x)

plt.plot(x,y)
plt.grid(which='both')
plt.yscale('log')
plt.xscale('log')

plt.show()
```

# Questão 8

Vamos precisar fazer 3 gráficos: 

1.  Peso $\times$ Passos $\rightarrow$ Papel Milimetrado

2.  log(Peso) $\times$ log(Passos) $\rightarrow$ Papel Milimetrado

3.  Peso $\times$ Passos $\rightarrow$ Papel Log

```python
massas = np.array([0.0114,0.0119,0.0324,0.3981,1.1491,9.5703])

passos = np.array([527.7, 543.3, 456.4, 278.5, 262.8, 208.3])
```

```python
fig, axs = plt.subplots(3,1, figsize=(5,11))
plt.subplots_adjust(hspace=.5)

# Plot 1
axs[0].set_title('Massa X Passos')
axs[0].plot(massas,passos)
axs[0].scatter(massas,passos)
axs[0].grid()

# Plot 2
axs[1].set_title('log(M) X log(P)')
axs[1].plot(np.log10(massas), np.log10(passos))
axs[1].scatter(np.log10(massas), np.log10(passos))
axs[1].grid()
axs[1].set_yticks([1,2,3])

# Plot 3
axs[2].set_title('Massa X Passos - Papel LOG')
axs[2].plot(massas, passos)
axs[2].scatter(massas, passos)
axs[2].set_yscale('log')
axs[2].set_xscale('log')
axs[2].grid(which='both')
axs[2].set_yticks([10**2, 10**3])
axs[2].set_xlim(left=10**-2)

plt.show()
# plt.savefig('Plots.jpg')
```

Agora que temos os gráficos, vamos precisar também das tabelas, como estamos usando Latex, podemos exportar no formato desejado:

## Log(Massa) X Log(Passos/Min)

```python
print(tabulate({'Massa (Kg)':np.log10(massas), 'Passos/min':np.log10(passos)}, tablefmt="latex", floatfmt=".1f", headers='keys'))
```
